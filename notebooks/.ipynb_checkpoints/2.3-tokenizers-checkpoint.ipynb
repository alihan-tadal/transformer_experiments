## Character-based

Bu yntem olduka basittir. Kelimeler yerine karakterlere ID atanr iki ana avantaj salar:
- Vocabulary boyutu olduka klr.
- Unknown token says olduka klr.



Fakat bu yntem kusursuz deildir ve kendi sorunlarn beraberinde getirir. Tokenizer kelimeler yerine karakterleri numaralandrd iin cmlenin temsili byk bir anlam tamayacaktr. nk karakterler ou dilde tek balarna bir anlam tamazlar.

Ayn zamanda bir cmleyi temsil etmek iin daha ok token'a ihtiya duyulacaktr. imdi, ilenilen iki yntemin de iyi ynlerini kullanan bir yaklamdan bahsedeceiz

## Subword tokenizaiton

Subword tokenization algoritmalar, sk kullanlan kelimelerin daha kk alt kelimelere blnmemesi, ancak nadir kelimelerin anlaml alt kelimelere ayrtrlmas gerektii ilkesine dayanr.

rnein "annoyingly" kelimesi nadir bir kelime olarak deerlendirilebilir ve "annoying" ve "ly" gibi alt kelimelere ayrtrlabilir. Bu iki kelimenin tekrarlanmas daha olaandr ve "annoyingly" kelimesinin anlam bu iki kelimenin bileimi ile elde edilebilir.



Bu yntem semantik olarak daha faydaldr nk "token" ve "ization" token'larn da anlaml hale getirir. Bunlara ek olarak vocabulary boyutu kelime tabanl tokenization'a gre ok daha kktr. Bylece daha byk bir vocabulary temsil edilebilir.

Bu tip tokenizer'lar __Trke__ gibi eklemeli dillerde kullanldr.

## Encoding

Kelimeleri saylara evirme ilemine __encoding__ denir ve iki aamda gerekleir. Bunlardan ilki tokenizasyondur ve bunu ID dnm takip eder.

Birinci adm aadaki gibi zetlenebilir:
```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")

sequence = "Using a Transformer network is simple"
tokens = tokenizer.tokenize(sequence)
print(tokens)
```
Bunun sonucunda aadaki kt elde edilir:
```python
['Using', 'a', 'transform', '##er', 'network', 'is', 'simple']
```
Kullanlan bir subword tokenizer olduu iin, ktda `##er` gibi subword'ler de yer alr. Bu token'lar sonraki aamada ID'lerine dntrr. Bu ilem aadaki gibi gerekletirilebilir:
```python
ids = tokenizer.convert_tokens_to_ids(tokens)
print(ids)
```

Bunun sonucunda aadaki kt elde edilir:
```python
[7993, 170, 11303, 1200, 2443, 1110, 3014]
```

Bu aamadan sonra elde edilen ID'ler, modellerde kullanlabilir.

## Decoding

Decode ilemi, yaplan ilemlerin tam tersine etir. Bir rnekle aklanabilir.
```python
decoded_string = tokenizer.decode([7993, 170, 11303, 1200, 2443, 1110, 3014])
print(decoded_string)
```
kt:
```python
'Using a Transformer network is simple'
```

Bu fonksiyon ID'leri token'a evirmek ile kalmaz, okunabilir bir cmle haline getirir. Bu zellik yeni metin oluturma ya da tahmin etme ilerinde olduka kullanldr.
